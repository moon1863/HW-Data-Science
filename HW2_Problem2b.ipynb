# Problem 2(b)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import sklearn
from sklearn.neighbors import KernelDensity
```

```python
np.random.seed(100)
X = np.random.normal(0, 1,10000)[:, np.newaxis]
```

```python
X.shape
```

```python
plt.hist(X, bins=50)
plt.show()
```

# Gaussian Kernel Function

```python
from numpy import arange
for h in arange (0.1,1.1,0.1):
    X_plot = np.linspace(-5, 5, 1000)[:, np.newaxis]
    KD = KernelDensity(kernel='gaussian', bandwidth= h).fit(X)
    log_dens = KD.score_samples(X_plot)
    fig, ax = plt.subplots()
    ax.plot(X_plot[:, 0], np.exp(log_dens), color = 'red')
    plt.title('Bandwidth={:.3f} & kernel distribution = Gaussian'.format(h))
    plt.hist(X, bins=50,density=True)
    plt.show
```

# Exponential Kernel Function

```python
from numpy import arange
for h in arange (0.1,1.1,0.1):
    X_plot = np.linspace(-5, 5, 1000)[:, np.newaxis]
    KD = KernelDensity(kernel='exponential', bandwidth= h).fit(X)
    log_dens = KD.score_samples(X_plot)
    fig, ax = plt.subplots()
    ax.plot(X_plot[:, 0], np.exp(log_dens), color = 'blue')
    plt.title('Bandwidth={:.3f} & kernel distribution = Exponential'.format(h))
    plt.hist(X, bins=50,density=True)
    plt.show
```

```python
from numpy import arange
for h in arange (0.1,1.1,0.1):
    X_plot = np.linspace(-5, 5, 1000)[:, np.newaxis]
    KD = KernelDensity(kernel='epanechnikov', bandwidth= h).fit(X)
    log_dens = KD.score_samples(X_plot)
    fig, ax = plt.subplots()
    ax.plot(X_plot[:, 0], np.exp(log_dens), color = 'magenta')
    plt.title('Bandwidth={:.3f} & kernel distribution = Epanechnikov'.format(h))
    plt.hist(X, bins=50,density=True)
    plt.show
```
